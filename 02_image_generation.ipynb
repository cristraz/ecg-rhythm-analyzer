## Tarea B1 - Parámetros del generador ecg-image-kit

| Parámetro | Qué hace | Valor que necesitamos |
|---|---|---|
| --num_leads | Número de derivaciones | 1 (solo DII) |
| --resolution | Tamaño de la imagen | 224 |
| --num_columns | Columnas del layout | 1 |
| --remove_lead_names | Quita los textos de la imagen | activado |
| --standard_grid_color | Color de la cuadrícula | rosa/rojo |
| --rotate | Rotación aleatoria | 0-3 grados |
| --noise | Ruido en la imagen | leve |
