## Tarea B1 - Parámetros del generador ecg-image-kit

| Parámetro | Qué hace | Valor que necesitamos |
|---|---|---|
| --num_leads | Número de derivaciones | 1 (solo DII) |
| --resolution | Tamaño de la imagen | 224 |
| --num_columns | Columnas del layout | 1 |
| --remove_lead_names | Quita los textos de la imagen | activado |
| --standard_grid_color | Color de la cuadrícula | rosa/rojo |
| --rotate | Rotación aleatoria | 0-3 grados |
| --noise | Ruido en la imagen | leve |


```python
import numpy as np
import os

# Crear carpeta de salida
os.makedirs('señales_prueba', exist_ok=True)

# Crear 10 señales falsas (ondas senoidales)
for i in range(10):
    t = np.linspace(0, 10, 5000)
    señal = np.sin(2 * np.pi * t) + 0.5 * np.sin(4 * np.pi * t)
    np.save(f'señales_prueba/señal_{i}.npy', señal)

print("✅ 10 señales creadas correctamente")
```

```python
import matplotlib.pyplot as plt

# Cargar y graficar la primera señal
señal = np.load('señales_prueba/señal_0.npy')

plt.figure(figsize=(12, 3))
plt.plot(señal)
plt.title('Señal de prueba')
plt.xlabel('Muestras')
plt.ylabel('Voltaje')
plt.grid(True)
plt.show()
```

```python
os.makedirs('señales_csv', exist_ok=True)

for i in range(10):
    señal = np.load(f'señales_prueba/señal_{i}.npy')
    # El generador espera 12 columnas (una por derivación)
    # Repetimos la misma señal 12 veces
    datos = np.column_stack([señal] * 12)
    np.savetxt(f'señales_csv/señal_{i}.csv', datos, delimiter=',')

print("✅ Archivos CSV creados correctamente")
```

```python
!python ecg-image-generator/gen_ecg_images_from_data_batch.py \
  -i señales_csv/ \
  -o imagenes_prueba/ \
  --num_leads 1 \
  --num_columns 1 \
  --remove_lead_names \
  -r 224
```

```python
import wfdb
import numpy as np
import os

os.makedirs('señales_wfdb', exist_ok=True)

for i in range(10):
    señal = np.load(f'señales_prueba/señal_{i}.npy')
    # Repetimos la señal 12 veces (una por derivación)
    datos = np.column_stack([señal] * 12)

    wfdb.wrsamp(
        f'señal_{i}',
        fs=500,
        units=['mV'] * 12,
        sig_name=['I','II','III','aVR','aVL','aVF','V1','V2','V3','V4','V5','V6'],
        p_signal=datos,
        write_dir='señales_wfdb/'
    )

print("✅ Archivos WFDB creados correctamente")
```

```python
!python ecg-image-generator/gen_ecg_images_from_data_batch.py \
  -i /content/ECG-Digitiser/señales_wfdb/ \
  -o /content/ECG-Digitiser/imagenes_prueba/ \
  --num_leads 1 \
  --num_columns 1 \
  --remove_lead_names \
  -r 224
```

```python
os.makedirs('wfdb_data', exist_ok=True)

for i in range(10):
    señal = np.load(f'señales_prueba/señal_{i}.npy')
    datos = np.column_stack([señal] * 12)

    wfdb.wrsamp(
        f'{str(i).zfill(5)}',
        fs=500,
        units=['mV'] * 12,
        sig_name=['I','II','III','aVR','aVL','aVF','V1','V2','V3','V4','V5','V6'],
        p_signal=datos,
        write_dir='wfdb_data/'
    )

print("✅ Archivos creados correctamente")
```
